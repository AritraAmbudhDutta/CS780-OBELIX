# Helper Functions [20 marks]

<a id="helper"></a>

Write all the helper functions that will be used for value-based and policy based algorithms below. In case you want to add more helper functions, please feel free to add.

````python
#Value Network
def createValueNetwork(inDim, outDim, action_size, hDim = [32,32], activation = F.relu):
    #this creates a Feed Forward Neural Network class and instantiates it and returns the class
    #the class should be derived from torch nn.Module and it should have init and forward method at the very least
    #the forward function should return q-value for each possible action

    #Your code goes in here
    class ValueNetwork(nn.Module):
        def __init__(self):
            super().__init__()
            self.activation = activation
            self.action_size = action_size

            input_dim = inDim if action_size is None else (inDim + action_size)
            self.layers = nn.ModuleList()
            prev = input_dim
            for h in hDim:
                self.layers.append(nn.Linear(prev, h))
                prev = h
            self.out = nn.Linear(prev, outDim)

        def forward(self, state, action=None):
            if self.action_size is None:
                x = state
            else:
                if action is None:
                    raise ValueError("action must be provided for Q-value network")
                x = torch.cat([state, action], dim=1)

            for layer in self.layers:
                x = self.activation(layer(x))
            return self.out(x)

    valueNetwork = ValueNetwork()
    return valueNetwork
    
````

````python
#Policy Network
def createPolicyNetwork(inDim, outDim, hDim = [32,32], activation = F.relu):
    #this creates a Feed Forward Neural Network class and instantiates it and returns the class
    #the class should be derived from torch nn.Module and it should have init and forward method at the very least
    #the forward function should return action logit vector
    #Your code goes in here
    class PolicyNetwork(nn.Module):
        def __init__(self):
            super(PolicyNetwork, self).__init__()
            self.activation = activation
            self.layers = nn.ModuleList()
            prevDim = inDim
            for h in hDim:
                self.layers.append(nn.Linear(prevDim, h))
                prevDim = h
            self.out = nn.Linear(prevDim, outDim)

        def forward(self, x):
            for layer in self.layers:
                x = self.activation(layer(x))
            return self.out(x)

    policyNetwork = PolicyNetwork()
    return policyNetwork
````

## ReplayBuffer [40 marks]

In next few cells, you will implement replaybuffer class.

This class creates a buffer for storing and retrieving experiences. This is a generic class and can be used
for different agents like NFQ, DQN, DDQN, PER_DDQN, etc.
Following are the methods for this class which are implemented in subsequent cells

```
class ReplayBuffer():
    def __init__(self, bufferSize, batch_size, seed)
    def store(self, state, action, reward, next_state, done)
    def sample(self, batchSize)
    def length(self)
```   

````python
from collections import deque
import random

class ReplayBuffer:
    def __init__(self, bufferSize, batch_size, seed):
        self.buffer = deque(maxlen=bufferSize)
        self.batch_size = batch_size
        random.seed(seed)

    def store(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batchSize):
        batch = random.sample(self.buffer, batchSize)
        return zip(*batch)

    def __len__(self):
        return len(self.buffer)

    def length(self):
        return len(self.buffer)
````

````python
# def __init__
class REINFORCE():
    def __init__(self, env, seed, gamma, optimizerFn, optimizerLR, MAX_TRAIN_EPISODES, MAX_EVAL_EPISODES, explorationStrategyTrainFn, explorationStrategyEvalFn):
        self.env = env
        self.seed = seed
        self.gamma = gamma
        self.optimizerFn = optimizerFn
        self.optimizerLR = optimizerLR
        self.MAX_TRAIN_EPISODES = MAX_TRAIN_EPISODES
        self.MAX_EVAL_EPISODES = MAX_EVAL_EPISODES
        self.explorationStrategyTrainFn = explorationStrategyTrainFn
        self.explorationStrategyEvalFn = explorationStrategyEvalFn
        
        torch.manual_seed(self.seed)
        np.random.seed(self.seed)
        
        self.inDim = np.prod(self.env.observation_space.shape)
        
        if isinstance(self.env.action_space, gym.spaces.Discrete):
            self.outDim = self.env.action_space.n
        else:
            self.outDim = self.env.action_space.shape[0] * 2
            
        self.policyNetwork = createPolicyNetwork(self.inDim, self.outDim).to(device)
        self.optimizer = self.optimizerFn(self.policyNetwork.parameters(), lr=self.optimizerLR)
        self.initBookKeeping()
````

````python
# def initBookKeeping
class REINFORCE(REINFORCE):
    def initBookKeeping(self):
        self.trainRewardsList = []
        self.trainTimeList = []
        self.evalRewardsList = []
        self.wallClockTimeList = []
        self.startTime = time.time()
        self.trainStartTime = time.time()
        self.episodeReward = 0
````

````python
# def performBookKeeping
class REINFORCE(REINFORCE):
    def performBookKeeping(self, train=True):
        if train:
            self.trainRewardsList.append(self.episodeReward)
            self.trainTimeList.append(time.time() - self.trainStartTime)
        else:
            self.evalRewardsList.append(self.episodeReward)
        self.wallClockTimeList.append(time.time() - self.startTime)
````

````python
# def runREINFORCE
class REINFORCE(REINFORCE):
    def runREINFORCE(self):
        self.trainAgent()
        finalEvalReward = self.evaluateAgent()
        return self.trainRewardsList, self.trainTimeList, self.evalRewardsList, self.wallClockTimeList, finalEvalReward
````

````python
# def trainAgent
class REINFORCE(REINFORCE):
    def trainAgent(self):
        for episode in range(self.MAX_TRAIN_EPISODES):
            state, _ = self.env.reset(seed=self.seed + episode)
            log_probs = []
            rewards = []
            self.episodeReward = 0
            done = False
            
            while not done:
                state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                action, log_prob = self.explorationStrategyTrainFn(self.policyNetwork, state_tensor)
                next_state, reward, terminated, truncated, _ = self.env.step(action)
                done = terminated or truncated
                
                log_probs.append(log_prob)
                rewards.append(reward)
                self.episodeReward += reward
                state = next_state
                
            self.performBookKeeping(train=True)
            self.trainPolicyNetwork((log_probs, rewards))
            self.evaluateAgent()
````

````python
# def trainPolicyNetwork
class REINFORCE(REINFORCE):
    def trainPolicyNetwork(self, experiences):
        log_probs, rewards = experiences
        returns = []
        G = 0
        for r in reversed(rewards):
            G = r + self.gamma * G
            returns.insert(0, G)
        
        returns = torch.tensor(returns).to(device)
        returns = (returns - returns.mean()) / (returns.std() + 1e-9)
        
        policy_loss = []
        for log_prob, G in zip(log_probs, returns):
            policy_loss.append(-log_prob * G)
            
        self.optimizer.zero_grad()
        policy_loss = torch.stack(policy_loss).sum()
        policy_loss.backward()
        self.optimizer.step()
````

````python
# def evaluateAgent
class REINFORCE(REINFORCE):
    def evaluateAgent(self):
        self.episodeReward = 0
        for episode in range(self.MAX_EVAL_EPISODES):
            state, _ = self.env.reset(seed=self.seed + 1000 + episode)
            done = False
            while not done:
                state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                action, _ = self.explorationStrategyEvalFn(self.policyNetwork, state_tensor)
                next_state, reward, terminated, truncated, _ = self.env.step(action)
                done = terminated or truncated
                self.episodeReward += reward
                state = next_state
        self.episodeReward /= self.MAX_EVAL_EPISODES
        self.performBookKeeping(train=False)
        return self.episodeReward
````

class VPG():

````python
# def __init__
class VPG():
    def __init__(self, env, seed, gamma, beta, optimizerFn, optimizerLR, MAX_TRAIN_EPISODES, MAX_EVAL_EPISODES, explorationStrategyTrainFn, explorationStrategyEvalFn):
        self.env = env
        self.seed = seed
        self.gamma = gamma
        self.beta = beta
        self.optimizerFn = optimizerFn
        self.optimizerLR = optimizerLR
        self.MAX_TRAIN_EPISODES = MAX_TRAIN_EPISODES
        self.MAX_EVAL_EPISODES = MAX_EVAL_EPISODES
        self.explorationStrategyTrainFn = explorationStrategyTrainFn
        self.explorationStrategyEvalFn = explorationStrategyEvalFn
        
        torch.manual_seed(self.seed)
        np.random.seed(self.seed)
        
        self.inDim = np.prod(self.env.observation_space.shape)
        
        if isinstance(self.env.action_space, gym.spaces.Discrete):
            self.outDim = self.env.action_space.n
        else:
            self.outDim = self.env.action_space.shape[0] * 2
            
        self.policyNetwork = createPolicyNetwork(self.inDim, self.outDim).to(device)
        self.valueNetwork = createValueNetwork(self.inDim, 1, None).to(device)
        
        self.optimizer = self.optimizerFn(list(self.policyNetwork.parameters()) + list(self.valueNetwork.parameters()), lr=self.optimizerLR)
        self.initBookKeeping()
````

````python
# def initBookKeeping
class VPG(VPG):
    def initBookKeeping(self):
        self.trainRewardsList = []
        self.trainTimeList = []
        self.evalRewardsList = []
        self.wallClockTimeList = []
        self.startTime = time.time()
        self.trainStartTime = time.time()
        self.episodeReward = 0
````

````python
# def performBookKeeping
class VPG(VPG):
    def performBookKeeping(self, train=True):
        if train:
            self.trainRewardsList.append(self.episodeReward)
            self.trainTimeList.append(time.time() - self.trainStartTime)
        else:
            self.evalRewardsList.append(self.episodeReward)
        self.wallClockTimeList.append(time.time() - self.startTime)
````

````python
# def runVPG
class VPG(VPG):
    def runVPG(self):
        self.trainAgent()
        finalEvalReward = self.evaluateAgent()
        return self.trainRewardsList, self.trainTimeList, self.evalRewardsList, self.wallClockTimeList, finalEvalReward
````

````python
# def trainAgent
class VPG(VPG):
    def trainAgent(self):
        for episode in range(self.MAX_TRAIN_EPISODES):
            state, _ = self.env.reset(seed=self.seed + episode)
            log_probs = []
            values = []
            rewards = []
            self.episodeReward = 0
            done = False
            
            while not done:
                state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                action, log_prob = self.explorationStrategyTrainFn(self.policyNetwork, state_tensor)
                value = self.valueNetwork(state_tensor)
                
                next_state, reward, terminated, truncated, _ = self.env.step(action)
                done = terminated or truncated
                
                log_probs.append(log_prob)
                values.append(value)
                rewards.append(reward)
                self.episodeReward += reward
                state = next_state
                
            self.performBookKeeping(train=True)
            self.trainPolicyNetwork((log_probs, values, rewards))
            self.evaluateAgent()
````

````python
# def trainPolicyNetwork
class VPG(VPG):
    def trainPolicyNetwork(self, experiences):
        log_probs, values, rewards = experiences
        returns = []
        G = 0
        for r in reversed(rewards):
            G = r + self.gamma * G
            returns.insert(0, G)
            
        returns = torch.FloatTensor(returns).to(device)
        returns = (returns - returns.mean()) / (returns.std() + 1e-9)
        
        policy_loss = []
        value_loss = []
        
        for log_prob, value, G in zip(log_probs, values, returns):
            advantage = G - value.item()
            policy_loss.append(-log_prob * advantage)
            value_loss.append(F.mse_loss(value, torch.FloatTensor([[G.item()]]).to(device)))
            
        self.optimizer.zero_grad()
        loss = torch.stack(policy_loss).sum() + torch.stack(value_loss).sum()
        loss.backward()
        self.optimizer.step()
````

````python
# def evaluateAgent
class VPG(VPG):
    def evaluateAgent(self):
        self.episodeReward = 0
        for episode in range(self.MAX_EVAL_EPISODES):
            state, _ = self.env.reset(seed=self.seed + 1000 + episode)
            done = False
            while not done:
                state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                action, _ = self.explorationStrategyEvalFn(self.policyNetwork, state_tensor)
                next_state, reward, terminated, truncated, _ = self.env.step(action)
                done = terminated or truncated
                self.episodeReward += reward
                state = next_state
        self.episodeReward /= self.MAX_EVAL_EPISODES
        self.performBookKeeping(train=False)
        return self.episodeReward
````

# Asynchronous Advantange Actor-Critic (A3C) [50 marks]
<a id="a3c"></a>

A3C (Mnih et al., 2016) was one of the first successful deep RL algorithms to train on CPU using parallel workers. Each worker maintains its own copy of the environment and policy, collects experience independently, and asynchronously pushes gradient updates to a shared global network.

**Key ideas:**
- Multiple worker processes each interact with their own environment copy.
- Each worker computes gradients locally and applies them to a **shared global network** asynchronously (no waiting for other workers).
- The asynchronous updates provide natural exploration diversity (different workers explore different parts of the state space).
- Uses n-step returns and advantage estimation similar to A2C.

**Update rule per worker:**
$$\theta \leftarrow \theta + \alpha \nabla_\theta \log \pi_\theta(a_t|s_t) \cdot A(s_t, a_t)$$

where the gradient is computed by the worker and applied to the shared $\theta$.

````python
# ======================================================================
# A3C Implementation
# ======================================================================

import torch.multiprocessing as mp
import torch.distributions as dist
import math

# A3C Hyperparameters
a3c_gym_id = "Hopper-v4"
a3c_learning_rate = 1e-4
a3c_gamma = 0.99
a3c_num_workers = 4                         # Number of asynchronous worker processes
a3c_num_steps = 5                           # n-step return length per worker
a3c_max_episodes_per_worker = 300
a3c_ent_coef = 0.01
a3c_vf_coef = 0.5
a3c_max_grad_norm = 0.5
````

````python
# ======================================================================
# SharedAdam: Adam optimizer with shared state across processes
# ======================================================================
# Standard Adam stores its internal state (moment estimates) per-process.
# For A3C, we need the optimizer state to be in shared memory so that
# all worker processes update the same momentum buffers.

class SharedAdam(optim.Adam):

    def __init__(self, params, lr=1e-3, betas=(0.9, 0.999), eps=1e-8):
        # Your code goes here
        super(SharedAdam, self).__init__(params, lr=lr, betas=betas, eps=eps)
        for group in self.param_groups:
            for p in group['params']:
                state = self.state[p]
                state['step'] = torch.zeros(1)
                state['exp_avg'] = torch.zeros_like(p.data)
                state['exp_avg_sq'] = torch.zeros_like(p.data)
                state['step'].share_memory_()
                state['exp_avg'].share_memory_()
                state['exp_avg_sq'].share_memory_()
````

````python
# ==============================================================
# A3C Agent
# ==============================================================

class A3CAgent(nn.Module):
    """
    A3C agent with separate actor and critic networks.
    Supports both discrete and continuous action spaces.
    """

    def __init__(self, obs_dim, act_dim, is_discrete=False):
        """
        Initialize actor and critic networks.
        obs_dim     : Dimension of observation space
        act_dim     : Dimension of action space
        is_discrete : Whether the action space is discrete
        """

        # Your code goes here
        super(A3CAgent, self).__init__()
        self.is_discrete = is_discrete
        
        self.critic = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )
        
        self.actor_mean = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, act_dim)
        )
        
        if not self.is_discrete:
            self.actor_logstd = nn.Parameter(torch.zeros(1, act_dim))

    def get_value(self, x):
        """
        Return the value estimate for a given state using the critic.
        """
        return self.critic(x)


    def get_action_and_value(self, x, action=None):
        """
        Compute action, log probability, entropy, and value.

        For discrete actions:
            create a categorical distribution from actor logits.

        For continuous actions:
            create a normal distribution using mean and std.

        If action is None, sample from the distribution.
        Return action, log probability, entropy, and value.
        """
        value = self.critic(x)
        mean = self.actor_mean(x)
        
        if self.is_discrete:
            distrib = dist.Categorical(logits=mean)
        else:
            std = self.actor_logstd.expand_as(mean).exp()
            distrib = dist.Normal(mean, std)
            
        if action is None:
            action = distrib.sample()
            
        log_prob = distrib.log_prob(action)
        
        if not self.is_discrete:
            log_prob = log_prob.sum(dim=-1)
            
        entropy = distrib.entropy()
        
        if not self.is_discrete:
            entropy = entropy.sum(dim=-1)
            
        return action, log_prob, entropy, value
````

````python
# A3C Worker (truly parallel via torch.multiprocessing)

def a3c_worker(worker_id, global_agent, global_optimizer, grad_lock,
               env_id, seed, num_episodes,
               gamma, num_steps, ent_coef, vf_coef, max_grad_norm,
               result_queue):
    """A3C worker running in a separate process."""
    env = gym.make(env_id)
    env.action_space.seed(seed + worker_id)
    torch.manual_seed(seed + worker_id)

    obs_dim = env.observation_space.shape[0]
    act_dim = env.action_space.shape[0]
    local_agent = A3CAgent(obs_dim, act_dim, is_discrete=False)

    for episode in range(num_episodes):
        local_agent.load_state_dict(global_agent.state_dict())
        state, _ = env.reset(seed=seed + worker_id + episode)
        done = False
        ep_reward = 0.0

        while not done:
            rewards, values, log_probs, entropies = [], [], [], []

            for _ in range(num_steps):
                state_tensor = torch.FloatTensor(state).unsqueeze(0)
                action, log_prob, entropy, value = local_agent.get_action_and_value(state_tensor)

                action_np = action.squeeze(0).detach().numpy()
                action_np = np.clip(action_np, env.action_space.low, env.action_space.high)

                next_state, reward, terminated, truncated, _ = env.step(action_np)
                done = terminated or truncated

                rewards.append(reward)
                values.append(value)
                log_probs.append(log_prob)
                entropies.append(entropy)

                ep_reward += reward
                state = next_state

                if done:
                    break

            R = torch.zeros(1, 1)
            if not done:
                with torch.no_grad():
                    state_tensor = torch.FloatTensor(state).unsqueeze(0)
                    R = local_agent.get_value(state_tensor)

            policy_loss = 0.0
            value_loss = 0.0
            entropy_loss = 0.0

            for i in reversed(range(len(rewards))):
                R = rewards[i] + gamma * R
                advantage = R - values[i]
                value_loss = value_loss + advantage.pow(2)
                policy_loss = policy_loss - log_probs[i] * advantage.detach()
                entropy_loss = entropy_loss - entropies[i]

            total_loss = policy_loss + vf_coef * value_loss + ent_coef * entropy_loss
            total_loss = total_loss.mean()

            local_agent.zero_grad()
            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(local_agent.parameters(), max_grad_norm)

            with grad_lock:
                global_optimizer.zero_grad()
                for local_param, global_param in zip(local_agent.parameters(), global_agent.parameters()):
                    global_param._grad = local_param.grad
                global_optimizer.step()

            local_agent.load_state_dict(global_agent.state_dict())

        result_queue.put((worker_id, episode, float(ep_reward)))

    env.close()
````

````python
# A3C Launcher — launches truly parallel worker processes
def train_a3c(env_id=a3c_gym_id, seed=1, num_workers=a3c_num_workers,
              episodes_per_worker=100,
              gamma=a3c_gamma, num_steps=a3c_num_steps, lr=a3c_learning_rate,
              ent_coef=a3c_ent_coef, vf_coef=a3c_vf_coef,
              max_grad_norm=a3c_max_grad_norm):
    """Launch A3C training with parallel worker processes."""

    ctx = mp.get_context("spawn")

    temp_env = gym.make(env_id)
    obs_dim = temp_env.observation_space.shape[0]
    act_dim = temp_env.action_space.shape[0]
    temp_env.close()

    global_agent = a3c_mod.A3CAgent(obs_dim, act_dim, is_discrete=False)
    global_agent.share_memory()

    global_optimizer = a3c_mod.SharedAdam(global_agent.parameters(), lr=lr)
    grad_lock = ctx.Lock()
    result_queue = ctx.Queue()

    processes = []
    for worker_id in range(num_workers):
        p = ctx.Process(
            target=a3c_mod.a3c_worker,
            args=(
                worker_id, global_agent, global_optimizer, grad_lock,
                env_id, seed, episodes_per_worker,
                gamma, num_steps, ent_coef, vf_coef, max_grad_norm,
                result_queue
            )
        )
        p.start()
        processes.append(p)

    results = []
    expected_results = num_workers * episodes_per_worker
    for _ in range(expected_results):
        results.append(result_queue.get())

    for p in processes:
        p.join()

    return results, global_agent
````

## A3C Questions [$10 \times 3 = 30$ marks]

### Q1. Why does A3C use asynchronous updates, and what benefit does this provide over synchronous (A2C) updates?

#### Ans:
##### a3c uses multiple workers that run their own environments and update a shared network independently. this gives better exploration because workers see different parts of the state space at the same time. it's better than a2c because it doesn't wait for all workers to finish a step, so it runs faster and never gets bottlenecked by the slowest worker.

### Q2. What is the "stale gradient" problem in A3C and how does it affect learning?

#### Ans:
##### stale gradients happen because a worker calculates its update using an older version of the global network. by the time it actually applies its update, other workers might have already changed the global weights. this makes the worker's update outdated, which adds random noise and can sometimes make the training process unstable.


### Q3. Compare A3C with DQN in terms of memory requirements and training paradigm.

#### Ans:
##### dqn needs a massive amount of memory because it saves past experiences in a huge replay buffer and trains off-policy by randomly sampling from it. a3c uses way less memory because it doesn't use a replay buffer at all. instead, it trains on-policy by using parallel workers that just gather fresh data, use it for a quick update, and throw it away.

# Advantange Actor-Critic (A2C) [50 marks]
<a id="a2c"></a>

A2C is the synchronous, deterministic variant of A3C. Instead of using asynchronous workers, A2C waits for all parallel workers to finish their rollouts before performing a single gradient update. This makes it simpler to implement and often equally effective.

**Key ideas:**
- Uses an Actor-Critic architecture: the actor outputs a policy $\pi(a|s)$ and the critic estimates $V(s)$.
- The advantage $A(s,a) = R_t - V(s_t)$ (or using GAE) is used to reduce variance in the policy gradient.
- Multiple environments run in parallel (vectorized) and all updates are synchronous.
- The loss combines: policy gradient loss, value loss, and an entropy bonus for exploration.

**Policy Gradient with Advantage:**
$$\nabla_\theta J(\theta) = \mathbb{E}\left[\nabla_\theta \log \pi_\theta(a_t|s_t) \cdot A(s_t, a_t)\right]$$

````python
# ======================================================================
# A2C Agent Network
# ======================================================================
# This class defines the architecture for an Advantage Actor-Critic (A2C) agent.
# A2C has two parts:
# 1. A critic that estimates the value of a state (V(s)).
# 2. An actor that outputs a probability distribution over actions.
# You will create separate networks for the actor and the critic.

class A2CAgent(nn.Module):
    def __init__(self, envs):
        # Initialize the A2C agent.
        # Tasks:
        # - Determine the dimensions of the observation space.
        # - Determine the number of possible actions.
        # - Create the critic network (takes state, outputs V(s)).
        # - Create the actor network (takes state, outputs action logits).

        # Your code goes here
        super().__init__()
        self.is_discrete = isinstance(envs.action_space, gym.spaces.Discrete)
        obs_dim = envs.observation_space.shape[0]
        
        if self.is_discrete:
            act_dim = envs.action_space.n
        else:
            act_dim = envs.action_space.shape[0]

        self.critic = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 1)
        )

        self.actor_mean = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, act_dim)
        )

        if not self.is_discrete:
            self.actor_logstd = nn.Parameter(torch.zeros(1, act_dim))

    def get_value(self, x):
        """Return the value of a state from the critic network."""
        # Your code goes here
        return self.critic(x)

    def get_action_and_value(self, x, action=None):
        """
        Get an action from the actor and value from the critic.
        Tasks:
        - Use the actor network to produce logits over actions.
        - Convert logits into a probability distribution.
        - Sample an action from the distribution (if action is not provided).
        - Compute the log probability of the chosen action.
        - Compute the entropy of the distribution (helps with exploration).
        - Get the value of the state from the critic.

        Returns: action, log_prob, entropy, value
        """
        # Your code goes here
        mean = self.actor_mean(x)
        if self.is_discrete:
            dist = torch.distributions.Categorical(logits=mean)
        else:
            action_logstd = self.actor_logstd.expand_as(mean)
            action_std = torch.exp(action_logstd)
            dist = torch.distributions.Normal(mean, action_std)

        if action is None:
            action = dist.sample()

        if self.is_discrete:
            log_prob = dist.log_prob(action)
            entropy = dist.entropy()
        else:
            log_prob = dist.log_prob(action).sum(1)
            entropy = dist.entropy().sum(1)

        return action, log_prob, entropy, self.critic(x)
````

````python
# ======================================================================
# A2C Worker + Training Loop (multiprocessing version)
# ======================================================================
# A2C is the SYNCHRONOUS counterpart to A3C:
#   - Worker processes each run their own environment and collect rollouts.
#   - After each rollout, workers WAIT at a barrier (synchronous).
#   - The main process computes GAE and does a SINGLE gradient update.
#   - A second barrier signals workers that weights have been updated.
# This barrier-based synchronisation is the key difference from A3C.
# ======================================================================

def a2c_worker(worker_id, global_agent,
               env_id, seed, num_steps, num_updates,
               shared_obs, shared_actions, shared_logprobs,
               shared_rewards, shared_dones, shared_values,
               shared_next_obs, shared_next_done,
               collect_barrier, update_barrier, ep_queue):
    """
    A2C worker process.  Collects rollouts into shared memory buffers,
    then waits at a barrier for the main process to do a centralised update.

    Unlike A3C, this worker NEVER computes gradients or calls optimizer.step().
    The main process owns the gradient update exclusively.
    """

    #Your code goes here
    env = gym.make(env_id)
    env.action_space.seed(seed + worker_id)
    obs, _ = env.reset(seed=seed + worker_id)
    done = False

    for update in range(num_updates):
        for step in range(num_steps):
            obs_tensor = torch.FloatTensor(obs).unsqueeze(0)
            with torch.no_grad():
                action, logprob, _, value = global_agent.get_action_and_value(obs_tensor)

            action_np = action.squeeze(0).numpy()
            next_obs, reward, terminated, truncated, _ = env.step(action_np)
            done = terminated or truncated

            shared_obs[step, worker_id] = obs_tensor.squeeze(0)
            shared_actions[step, worker_id] = action.squeeze(0)
            shared_logprobs[step, worker_id] = logprob.squeeze(0)
            shared_rewards[step, worker_id] = reward
            shared_dones[step, worker_id] = done
            shared_values[step, worker_id] = value.squeeze(0)

            obs = next_obs

            if done:
                obs, _ = env.reset()
                ep_queue.put((update, reward))

        shared_next_obs[worker_id] = torch.FloatTensor(obs)
        shared_next_done[worker_id] = done

        collect_barrier.wait()
        update_barrier.wait()

    env.close()

def train_a2c(env_id=a2c_gym_id, seed=a2c_seed, total_timesteps=a2c_total_timesteps,
              num_envs=a2c_num_envs, num_steps=a2c_num_steps, lr=a2c_learning_rate,
              gamma=a2c_gamma, gae_lambda=a2c_gae_lambda, use_gae=a2c_use_gae,
              ent_coef=a2c_ent_coef, vf_coef=a2c_vf_coef,
              max_grad_norm=a2c_max_grad_norm, norm_adv=a2c_norm_adv):
    """
    A2C training with torch.multiprocessing.

    - num_envs worker processes each run their own environment.
    - Workers collect num_steps of experience into shared memory buffers.
    - A barrier synchronises: all workers must finish before the update.
    - The main process computes GAE advantages and does ONE gradient step.
    - A second barrier signals workers that weights have been updated.
    """

    ctx = mp.get_context("spawn")

    temp_env = gym.make(env_id)
    obs_dim = temp_env.observation_space.shape[0]
    if isinstance(temp_env.action_space, gym.spaces.Discrete):
        act_dim = 1
        action_dtype = torch.long
    else:
        act_dim = temp_env.action_space.shape[0]
        action_dtype = torch.float
    temp_env.close()

    global_agent = a2c_mod.A2CAgent(gym.make(env_id))
    global_agent.share_memory()
    optimizer = optim.RMSprop(global_agent.parameters(), lr=lr, eps=1e-5)

    shared_obs = torch.zeros((num_steps, num_envs, obs_dim)).share_memory_()
    shared_actions = torch.zeros((num_steps, num_envs, act_dim), dtype=action_dtype).share_memory_()
    shared_logprobs = torch.zeros((num_steps, num_envs)).share_memory_()
    shared_rewards = torch.zeros((num_steps, num_envs)).share_memory_()
    shared_dones = torch.zeros((num_steps, num_envs)).share_memory_()
    shared_values = torch.zeros((num_steps, num_envs)).share_memory_()
    shared_next_obs = torch.zeros((num_envs, obs_dim)).share_memory_()
    shared_next_done = torch.zeros(num_envs).share_memory_()

    collect_barrier = ctx.Barrier(num_envs + 1)
    update_barrier = ctx.Barrier(num_envs + 1)
    ep_queue = ctx.Queue()

    batch_size = num_envs * num_steps
    num_updates = total_timesteps // batch_size

    processes = []
    for worker_id in range(num_envs):
        p = ctx.Process(target=a2c_mod.a2c_worker, args=(
            worker_id, global_agent, env_id, seed, num_steps, num_updates,
            shared_obs, shared_actions, shared_logprobs,
            shared_rewards, shared_dones, shared_values,
            shared_next_obs, shared_next_done,
            collect_barrier, update_barrier, ep_queue
        ))
        p.start()
        processes.append(p)

    results = []

    for update in range(num_updates):
        collect_barrier.wait()

        with torch.no_grad():
            next_value = global_agent.get_value(shared_next_obs).squeeze(1)
            advantages = torch.zeros_like(shared_rewards)
            if use_gae:
                lastgaelam = 0
                for t in reversed(range(num_steps)):
                    if t == num_steps - 1:
                        nextnonterminal = 1.0 - shared_next_done
                        nextvalues = next_value
                    else:
                        nextnonterminal = 1.0 - shared_dones[t + 1]
                        nextvalues = shared_values[t + 1]
                    delta = shared_rewards[t] + gamma * nextvalues * nextnonterminal - shared_values[t]
                    advantages[t] = lastgaelam = delta + gamma * gae_lambda * nextnonterminal * lastgaelam
                returns = advantages + shared_values
            else:
                returns = torch.zeros_like(shared_rewards)
                for t in reversed(range(num_steps)):
                    if t == num_steps - 1:
                        nextnonterminal = 1.0 - shared_next_done
                        next_return = next_value
                    else:
                        nextnonterminal = 1.0 - shared_dones[t + 1]
                        next_return = returns[t + 1]
                    returns[t] = shared_rewards[t] + gamma * nextnonterminal * next_return
                advantages = returns - shared_values

        b_obs = shared_obs.view(-1, obs_dim)
        b_actions = shared_actions.view(-1, act_dim)
        if action_dtype == torch.long:
            b_actions = b_actions.squeeze(-1)
        b_advantages = advantages.view(-1)
        b_returns = returns.view(-1)

        if norm_adv:
            b_advantages = (b_advantages - b_advantages.mean()) / (b_advantages.std() + 1e-8)

        _, newlogprob, entropy, newvalue = global_agent.get_action_and_value(b_obs, b_actions)

        pg_loss = -(b_advantages * newlogprob).mean()
        v_loss = F.mse_loss(newvalue.view(-1), b_returns)
        entropy_loss = entropy.mean()

        loss = pg_loss - ent_coef * entropy_loss + v_loss * vf_coef

        optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(global_agent.parameters(), max_grad_norm)
        optimizer.step()

        while not ep_queue.empty():
            results.append(ep_queue.get())

        update_barrier.wait()

    for p in processes:
        p.join()

    return results
````

# Generalized Advantage Estimation (GAE) [50 marks]
<a id="gae"></a>

GAE (Schulman et al., 2016) provides a principled way to balance bias and variance in advantage estimation through a single parameter $\lambda \in [0, 1]$.

**The core idea:**

The 1-step advantage (TD error) is: $\delta_t^V = r_t + \gamma V(s_{t+1}) - V(s_t)$

The GAE advantage is an exponentially-weighted average of multi-step advantages:
$$\hat{A}_t^{\text{GAE}(\gamma, \lambda)} = \sum_{l=0}^{\infty} (\gamma \lambda)^l \delta_{t+l}^V$$

This can be computed recursively:
$$\hat{A}_t = \delta_t + \gamma \lambda \hat{A}_{t+1}$$

**Special cases:**
- $\lambda = 0$: GAE reduces to the 1-step TD advantage $\delta_t^V$ → **low variance, high bias**
- $\lambda = 1$: GAE reduces to the full Monte Carlo advantage → **high variance, low bias**
- $0 < \lambda < 1$: Smooth interpolation between the two extremes

Below we implement a standalone GAE module and demonstrate how different values of $\lambda$ affect learning performance.


````python
# ======================================================================
# GAE Implementation (standalone module)
# ======================================================================

def compute_gae(rewards, values, dones, next_value, gamma, gae_lambda):
    """
    Computes Generalized Advantage Estimation.

    Args:
        rewards    : tensor of shape (T, num_envs) - rewards at each timestep
        values     : tensor of shape (T, num_envs) - V(s_t) predictions
        dones      : tensor of shape (T, num_envs) - done flags
        next_value : tensor of shape (1, num_envs) - V(s_{T}) bootstrap value
        gamma      : discount factor
        gae_lambda : GAE lambda parameter (0=TD, 1=MC)

    Returns:
        advantages : tensor of shape (T, num_envs)
        returns    : tensor of shape (T, num_envs)
    """

    #Your code goes here
    T = rewards.size(0)
    advantages = torch.zeros_like(rewards)
    lastgaelam = 0
    for t in reversed(range(T)):
        if t == T - 1:
            nextnonterminal = 1.0 - dones[t]
            nextvalues = next_value
        else:
            nextnonterminal = 1.0 - dones[t+1]
            nextvalues = values[t+1]
        delta = rewards[t] + gamma * nextvalues * nextnonterminal - values[t]
        advantages[t] = lastgaelam = delta + gamma * gae_lambda * nextnonterminal * lastgaelam
    returns = advantages + values
    return advantages, returns
````

````python
class DDPG(DDPG):
    def updateNetworks(self, onlineNet, targetNet, tau):
        #this function updates the onlineNetwork with the target network
        #
        # Your code goes in here
        #
        for target_param, online_param in zip(targetNet.parameters(), onlineNet.parameters()):
            target_param.data.copy_(tau * online_param.data + (1.0 - tau) * target_param.data)
````

````python
class DDPG(DDPG):
    def gaussianStrategy (self, net , s , envActionRange , noiseScaleRatio ,
        explorationMax = True ):
        #this function sets the scale of exploration then add the noise of this scale to the greedy action
        #and clips it within the range

        #Your code here
        with torch.no_grad():
            action = net(s).cpu().numpy()[0]
        
        noise = np.random.normal(0, noiseScaleRatio * (envActionRange[1] - envActionRange[0]), size=action.shape)
        action = action + noise
        action = np.clip(action, envActionRange[0], envActionRange[1])
        return action
````

# Soft Actor-Critic (SAC) [ 50 marks ]

SAC is an off-policy maximum entropy deep RL algorithm for continuous action spaces. It maintains two online Q-networks and two target Q-networks to mitigate overestimation bias, and uses a squashed Gaussian policy with automatic entropy tuning.

### Key Ideas

**1. Stochastic Policy (Squashed Gaussian)**  
The network learns $\mu$ and $\sigma$; samples $u \sim \mathcal{N}(\mu, \sigma)$, applies $a = \tanh(u)$, then rescales to the environment action range.

**2. Log-probability correction for tanh squashing:**

$$\log \pi(a \mid s) = \log p(u \mid s) - \sum_{i=0}^{D} \log(1 - \tanh^2(u_i))$$

**3. Policy Loss (maximize Q, maximize entropy):**

$$J_\pi(\phi) = -\, \mathbb{E}_{s \sim \mathcal{U}(D),\, \hat{a} \sim \pi} \left[ \min_k Q(s, \hat{a};\, \theta^k) - \alpha \log \pi(\hat{a} \mid s;\, \phi) \right]$$

**4. Automatic Entropy Tuning (learn $\alpha$):**

$$J(\alpha) = -\, \mathbb{E}_{s \sim \mathcal{U}(D),\, \hat{a} \sim \pi} \left[ \alpha \left( \mathcal{H} + \log \pi(\hat{a} \mid s;\, \phi) \right) \right]$$

where $\mathcal{H} = -|\mathcal{A}| = -\prod \text{shape}(a_\text{env})$

**5. Soft Bellman Backup (critic target):**

$$\text{target\_q} = r + \gamma \cdot \left( \min_k Q_{\text{target},k}(s', a') - \alpha \log \pi(a' \mid s') \right) \cdot (1 - \text{done})$$

**6. Soft (Polyak) Target Update:**

$$\theta_{\text{target}} \leftarrow (1 - \tau)\, \theta_{\text{target}} + \tau\, \theta_{\text{online}}$$

````python
class ValueNetwork(nn.Module):

    def __init__(self, stateDim, actionDim, hiddenDims):
        """
        Args:
            stateDim (int): dimension of state space
            actionDim (int): dimension of action space
            hiddenDims (list): hidden layer sizes

        Output:
            Initializes Q-network Q(s,a)
        """
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(stateDim + actionDim, hiddenDims[0]),
            nn.ReLU(),
            nn.Linear(hiddenDims[0], hiddenDims[1]),
            nn.ReLU(),
            nn.Linear(hiddenDims[1], 1)
        )

    def forward(self, state, action):
        """
        Args:
            state (tensor): shape (batch_size, stateDim)
            action (tensor): shape (batch_size, actionDim)

        Returns:
            q_value (tensor): predicted Q(s,a), shape (batch_size,1)
        """
        x = torch.cat([state, action], dim=1)
        return self.net(x)
````

````python
class PolicyNetwork(nn.Module):

    def __init__(self, stateDim, actionRange, hiddenDims):
        """
        Args:
            stateDim (int): state dimension
            actionRange (tuple): (action_low, action_high)
            hiddenDims (list): hidden layer sizes

        Output:
            Initializes stochastic policy network
        """
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(stateDim, hiddenDims[0]),
            nn.ReLU(),
            nn.Linear(hiddenDims[0], hiddenDims[1]),
            nn.ReLU()
        )
        actionDim = actionRange[1].shape[0]
        self.mean_linear = nn.Linear(hiddenDims[1], actionDim)
        self.log_std_linear = nn.Linear(hiddenDims[1], actionDim)
        
        self.action_scale = torch.FloatTensor((actionRange[1] - actionRange[0]) / 2.0).to(device)
        self.action_bias = torch.FloatTensor((actionRange[1] + actionRange[0]) / 2.0).to(device)

    def forward(self, state):
        """
        Args:
            state (tensor): input states

        Returns:
            action_env (tensor): sampled action in env range
            action_greedy (tensor): deterministic action
            log_prob (tensor): log probability of action
        """
        x = self.net(state)
        mean = self.mean_linear(x)
        log_std = self.log_std_linear(x)
        log_std = torch.clamp(log_std, min=-20, max=2)
        std = log_std.exp()
        
        normal = Normal(mean, std)
        x_t = normal.rsample()
        y_t = torch.tanh(x_t)
        action_env = y_t * self.action_scale + self.action_bias
        
        log_prob = normal.log_prob(x_t)
        log_prob -= torch.log(self.action_scale * (1 - y_t.pow(2)) + 1e-6)
        log_prob = log_prob.sum(1, keepdim=True)
        
        mean_tanh = torch.tanh(mean)
        action_greedy = mean_tanh * self.action_scale + self.action_bias
        
        return action_env, action_greedy, log_prob

    def rescale(self, action):
        """
        Args:
            action (tensor): tanh output in [-1,1]

        Returns:
            scaled_action (tensor): action in environment range
        """
        return action * self.action_scale + self.action_bias
````

````python
class SAC:

    def __init__(self, env, gamma, tau, bufferSize, batchSize,
                 entropyLR=3e-4, updateFrequency=1,
                 policyOptimizerFn=torch.optim.Adam,
                 valueOptimizerFn_1=torch.optim.Adam,
                 valueOptimizerFn_2=torch.optim.Adam,
                 policyOptimizerLR=3e-4,
                 valueOptimizerLR=3e-4,
                 alphaOptimizerFn=torch.optim.Adam,
                 MAX_TRAIN_EPISODES=120,
                 MAX_EVAL_EPISODE=1,
                 hDims=[256, 256],
                 minSamples=1000):
        """
        Args:
            env: gym environment
            gamma (float): discount factor
            tau (float): target update rate
            bufferSize (int): replay buffer size
            batchSize (int): training batch size
        """
        self.env = env
        self.gamma = gamma
        self.tau = tau
        self.batchSize = batchSize
        self.updateFrequency = updateFrequency
        self.MAX_TRAIN_EPISODES = MAX_TRAIN_EPISODES
        self.MAX_EVAL_EPISODE = MAX_EVAL_EPISODE
        self.minSamples = minSamples

        self.stateDim = env.observation_space.shape[0]
        self.actionDim = env.action_space.shape[0]
        self.actionRange = (env.action_space.low, env.action_space.high)
        
        self.target_entropy = -torch.prod(torch.Tensor(env.action_space.shape).to(device)).item()
        self.log_alpha = torch.zeros(1, requires_grad=True, device=device)
        self.alpha_optimizer = alphaOptimizerFn([self.log_alpha], lr=entropyLR)
        
        self.policy = PolicyNetwork(self.stateDim, self.actionRange, hDims).to(device)
        self.policy_optimizer = policyOptimizerFn(self.policy.parameters(), lr=policyOptimizerLR)
        
        self.critic1 = ValueNetwork(self.stateDim, self.actionDim, hDims).to(device)
        self.critic2 = ValueNetwork(self.stateDim, self.actionDim, hDims).to(device)
        self.critic1_target = ValueNetwork(self.stateDim, self.actionDim, hDims).to(device)
        self.critic2_target = ValueNetwork(self.stateDim, self.actionDim, hDims).to(device)
        
        self.critic1_target.load_state_dict(self.critic1.state_dict())
        self.critic2_target.load_state_dict(self.critic2.state_dict())
        
        self.critic1_optimizer = valueOptimizerFn_1(self.critic1.parameters(), lr=valueOptimizerLR)
        self.critic2_optimizer = valueOptimizerFn_2(self.critic2.parameters(), lr=valueOptimizerLR)
        
        self.replayBuffer = ReplayBuffer(bufferSize, self.batchSize, 0)

    def selectAction(self, state):
        """
        Args:
            state (array): current state

        Returns:
            action (array): action sampled from policy
        """
        state = torch.FloatTensor(state).unsqueeze(0).to(device)
        with torch.no_grad():
            action_env, _, _ = self.policy(state)
        return action_env.squeeze(0).cpu().numpy()

    def trainNetworks(self, batch):
        """
        Args:
            batch: sampled transitions

        Updates:
            critics, policy, entropy coefficient
        """
        states, actions, rewards, next_states, dones = batch
        
        states = torch.FloatTensor(np.array(list(states))).to(device)
        actions = torch.FloatTensor(np.array(list(actions))).to(device)
        rewards = torch.FloatTensor(np.array(list(rewards))).unsqueeze(1).to(device)
        next_states = torch.FloatTensor(np.array(list(next_states))).to(device)
        dones = torch.FloatTensor(np.array(list(dones))).unsqueeze(1).to(device)
        
        alpha = self.log_alpha.exp()
        
        with torch.no_grad():
            next_actions, _, next_log_probs = self.policy(next_states)
            target_q1 = self.critic1_target(next_states, next_actions)
            target_q2 = self.critic2_target(next_states, next_actions)
            target_q = torch.min(target_q1, target_q2) - alpha * next_log_probs
            y = rewards + self.gamma * (1 - dones) * target_q
            
        q1 = self.critic1(states, actions)
        q2 = self.critic2(states, actions)
        
        critic1_loss = F.mse_loss(q1, y)
        critic2_loss = F.mse_loss(q2, y)
        
        self.critic1_optimizer.zero_grad()
        critic1_loss.backward()
        self.critic1_optimizer.step()
        
        self.critic2_optimizer.zero_grad()
        critic2_loss.backward()
        self.critic2_optimizer.step()
        
        new_actions, _, log_probs = self.policy(states)
        q1_new = self.critic1(states, new_actions)
        q2_new = self.critic2(states, new_actions)
        q_new = torch.min(q1_new, q2_new)
        
        policy_loss = (alpha * log_probs - q_new).mean()
        
        self.policy_optimizer.zero_grad()
        policy_loss.backward()
        self.policy_optimizer.step()
        
        alpha_loss = -(self.log_alpha * (log_probs + self.target_entropy).detach()).mean()
        
        self.alpha_optimizer.zero_grad()
        alpha_loss.backward()
        self.alpha_optimizer.step()

    def updateTargetNetworks(self):
        """
        Performs Polyak update of target networks
        """
        for target_param, param in zip(self.critic1_target.parameters(), self.critic1.parameters()):
            target_param.data.copy_(self.tau * param.data + (1.0 - self.tau) * target_param.data)
        for target_param, param in zip(self.critic2_target.parameters(), self.critic2.parameters()):
            target_param.data.copy_(self.tau * param.data + (1.0 - self.tau) * target_param.data)

    def trainAgent(self):
        """
        Runs training episodes

        Returns:
            trainRewards (list): reward per episode
        """
        trainRewards = []
        trainTimeList = []
        startTime = time.time()
        total_steps = 0
        
        for episode in range(self.MAX_TRAIN_EPISODES):
            state, _ = self.env.reset()
            episode_reward = 0
            done = False
            
            while not done:
                if len(self.replayBuffer) < self.minSamples:
                    action = self.env.action_space.sample()
                else:
                    action = self.selectAction(state)
                    
                next_state, reward, terminated, truncated, _ = self.env.step(action)
                done = terminated or truncated
                
                self.replayBuffer.store(state, action, reward, next_state, done)
                
                if len(self.replayBuffer) >= self.batchSize and total_steps % self.updateFrequency == 0:
                    batch = self.replayBuffer.sample(self.batchSize)
                    self.trainNetworks(batch)
                    self.updateTargetNetworks()
                    
                state = next_state
                episode_reward += reward
                total_steps += 1
                
            trainRewards.append(episode_reward)
            trainTimeList.append(time.time() - startTime)
            
        return trainRewards, trainTimeList

    def evaluateAgent(self):
        """
        Returns:
            mean_reward (float): evaluation performance
        """
        eval_rewards = []
        for _ in range(self.MAX_EVAL_EPISODE):
            state, _ = self.env.reset()
            done = False
            ep_reward = 0
            while not done:
                state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                with torch.no_grad():
                    _, action_greedy, _ = self.policy(state_tensor)
                action = action_greedy.squeeze(0).cpu().numpy()
                next_state, reward, terminated, truncated, _ = self.env.step(action)
                done = terminated or truncated
                ep_reward += reward
                state = next_state
            eval_rewards.append(ep_reward)
        return np.mean(eval_rewards)

    def runSAC(self):
        """
        Runs full training pipeline

        Returns:
            train_rewards, eval_score, training_time
        """
        train_rewards, train_time = self.trainAgent()
        eval_score = self.evaluateAgent()
        return train_rewards, eval_score, train_time, None
````

````python
# PPO Agent (takes raw dims, works in any process)
class PPOAgent(nn.Module):
    """
    PPO agent containing separate actor and critic networks.
    Supports both discrete and continuous action spaces.
    """

    def __init__(self, obs_dim, act_dim, is_discrete=False,
                 action_high=None, action_low=None):
        """
        Initialize the PPO agent.

        Args:
            obs_dim (int): Dimension of the observation/state space.
            act_dim (int): Dimension of the action space.
            is_discrete (bool): True if the environment has a discrete
                                action space, otherwise False.
            action_high (array-like, optional): Upper bound of the
                                continuous action space.
            action_low (array-like, optional): Lower bound of the
                                continuous action space.

        Expected Output:
            Initializes actor and critic neural networks along with
            parameters required for action scaling in continuous spaces.
        """

        #Your Code goes here
        super().__init__()
        self.is_discrete = is_discrete
        
        self.critic = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 1)
        )
        
        self.actor_mean = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, act_dim)
        )
        
        if not self.is_discrete:
            self.actor_logstd = nn.Parameter(torch.zeros(1, act_dim))

    def get_value(self, x):
        """
        Compute the value estimate for a given state.

        Args:
            x (Tensor): Input state tensor of shape (batch_size, obs_dim).

        Expected Output:
            Tensor containing the predicted state value V(s)
            from the critic network.
        """
        return self.critic(x)



    def get_action_and_value(self, x, action=None):
        """
        Compute action, log probability, entropy, and state value.

        Args:
            x (Tensor): Input state tensor of shape (batch_size, obs_dim).
            action (Tensor, optional): Specific action for which the
                                       log probability should be computed.
                                       If None, the function samples an action
                                       from the policy distribution.

        Expected Output:
            action (Tensor): Sampled or provided action.
            log_prob (Tensor): Log probability of the selected action.
            entropy (Tensor): Entropy of the action distribution
                              (used for exploration).
            value (Tensor): Predicted state value from the critic.
        """
        mean = self.actor_mean(x)
        if self.is_discrete:
            dist = torch.distributions.Categorical(logits=mean)
        else:
            action_logstd = self.actor_logstd.expand_as(mean)
            action_std = torch.exp(action_logstd)
            dist = torch.distributions.Normal(mean, action_std)

        if action is None:
            action = dist.sample()

        if self.is_discrete:
            log_prob = dist.log_prob(action)
            entropy = dist.entropy()
        else:
            log_prob = dist.log_prob(action).sum(1)
            entropy = dist.entropy().sum(1)

        return action, log_prob, entropy, self.critic(x)
````

````python
# PPO Worker + Training Loop (Multiprocessing Version)
def ppo_worker(worker_id, global_agent,
               env_id, seed, num_steps, num_updates,
               shared_obs, shared_actions, shared_logprobs,
               shared_rewards, shared_dones, shared_values,
               shared_next_obs, shared_next_done,
               collect_barrier, update_barrier, ep_queue):
    """
    PPO worker process responsible for collecting rollouts.

    Args:
        worker_id        : Unique ID of the worker process.
        global_agent     : Shared PPO policy network.
        env_id           : Gym environment name.
        seed             : Random seed for reproducibility.
        num_steps        : Number of steps collected per rollout.
        num_updates      : Number of PPO updates to perform.

        shared_obs       : Shared buffer storing observations.
        shared_actions   : Shared buffer storing actions.
        shared_logprobs  : Shared buffer storing log probabilities.
        shared_rewards   : Shared buffer storing rewards.
        shared_dones     : Shared buffer storing termination flags.
        shared_values    : Shared buffer storing value predictions.

        shared_next_obs  : Shared buffer storing next observations.
        shared_next_done : Shared buffer storing next done flag.

        collect_barrier  : Synchronization barrier for rollout collection.
        update_barrier   : Synchronization barrier after PPO update.
        ep_queue         : Queue used to send episode rewards to main process.

    Expected Output:
        This function does not return values directly.
        Instead, it stores rollout data in shared buffers and
        sends episode rewards to the main process through ep_queue.
    """

    env = gym.make(env_id)
    env.action_space.seed(seed + worker_id)
    obs, _ = env.reset(seed=seed + worker_id)
    done = False
    ep_return = 0.0

    is_discrete = isinstance(env.action_space, gym.spaces.Discrete)
    if not is_discrete:
        action_low = torch.tensor(env.action_space.low, dtype=torch.float32)
        action_high = torch.tensor(env.action_space.high, dtype=torch.float32)

    for update in range(num_updates):
        for step in range(num_steps):
            obs_tensor = torch.FloatTensor(obs).unsqueeze(0)
            with torch.no_grad():
                action, _, _, value = global_agent.get_action_and_value(obs_tensor)

                if is_discrete:
                    action_to_store = action.long()
                    action_env = int(action_to_store.item())
                else:
                    action_to_store = torch.max(torch.min(action, action_high.unsqueeze(0)), action_low.unsqueeze(0))
                    action_env = action_to_store.squeeze(0).numpy()

                _, logprob, _, value = global_agent.get_action_and_value(obs_tensor, action_to_store)

            next_obs, reward, terminated, truncated, _ = env.step(action_env)
            done = terminated or truncated
            ep_return += reward

            shared_obs[step, worker_id] = obs_tensor.squeeze(0)
            shared_actions[step, worker_id] = action_to_store.squeeze(0)
            shared_logprobs[step, worker_id] = logprob.squeeze(0)
            shared_rewards[step, worker_id] = reward
            shared_dones[step, worker_id] = done
            shared_values[step, worker_id] = value.squeeze(0)

            obs = next_obs

            if done:
                ep_queue.put((update, float(ep_return)))
                obs, _ = env.reset()
                ep_return = 0.0

        shared_next_obs[worker_id] = torch.FloatTensor(obs)
        shared_next_done[worker_id] = done

        collect_barrier.wait()
        update_barrier.wait()

    env.close()
````

````python
def train_ppo(env_id=ppo_gym_id, seed=ppo_seed,
              total_timesteps=ppo_total_timesteps,
              num_workers=ppo_num_workers, num_steps=ppo_num_steps,
              lr=ppo_learning_rate, gamma=ppo_gamma,
              gae_lambda=ppo_gae_lambda, update_epochs=ppo_update_epochs,
              num_minibatches=ppo_num_minibatches,
              clip_coef=ppo_clip_coef, clip_vloss=ppo_clip_vloss,
              ent_coef=ppo_ent_coef, vf_coef=ppo_vf_coef,
              max_grad_norm=ppo_max_grad_norm, norm_adv=ppo_norm_adv,
              anneal_lr=ppo_anneal_lr, target_kl=ppo_target_kl):
    """
    PPO training loop using multiprocessing workers.

    Args:
        env_id           : Gym environment used for training.
        seed             : Random seed.
        total_timesteps  : Total number of environment interactions.
        num_workers      : Number of parallel workers collecting data.
        num_steps        : Number of rollout steps per worker.

        lr               : Learning rate for optimizer.
        gamma            : Discount factor for future rewards.
        gae_lambda       : Lambda parameter for GAE.

        update_epochs    : Number of PPO update passes per batch.
        num_minibatches  : Number of minibatches for SGD.

        clip_coef        : PPO clipping coefficient.
        clip_vloss       : Whether value loss clipping is used.

        ent_coef         : Entropy coefficient for exploration.
        vf_coef          : Value loss coefficient.
        max_grad_norm    : Maximum gradient norm for clipping.

        norm_adv         : Whether to normalize advantages.
        anneal_lr        : Whether learning rate is annealed.
        target_kl        : Optional KL divergence stopping threshold.

    Expected Output:
        results      : List containing tuples (global_step, episode_reward)
                       used to plot the learning curve.
        global_agent : The trained PPO agent.
    """
    ctx = mp.get_context("spawn")

    temp_env = gym.make(env_id)
    obs_dim = temp_env.observation_space.shape[0]
    is_discrete = isinstance(temp_env.action_space, gym.spaces.Discrete)

    if is_discrete:
        act_dim = temp_env.action_space.n
        action_dtype = torch.long
    else:
        act_dim = temp_env.action_space.shape[0]
        action_dtype = torch.float
    temp_env.close()

    global_agent = ppo_mod.PPOAgent(obs_dim, act_dim, is_discrete)
    global_agent.share_memory()
    optimizer = optim.Adam(global_agent.parameters(), lr=lr, eps=1e-5)

    shared_obs = torch.zeros((num_steps, num_workers, obs_dim)).share_memory_()
    shared_actions = torch.zeros((num_steps, num_workers, 1 if is_discrete else act_dim), dtype=action_dtype).share_memory_()
    shared_logprobs = torch.zeros((num_steps, num_workers)).share_memory_()
    shared_rewards = torch.zeros((num_steps, num_workers)).share_memory_()
    shared_dones = torch.zeros((num_steps, num_workers)).share_memory_()
    shared_values = torch.zeros((num_steps, num_workers)).share_memory_()
    shared_next_obs = torch.zeros((num_workers, obs_dim)).share_memory_()
    shared_next_done = torch.zeros(num_workers).share_memory_()

    collect_barrier = ctx.Barrier(num_workers + 1)
    update_barrier = ctx.Barrier(num_workers + 1)
    ep_queue = ctx.Queue()

    batch_size = num_workers * num_steps
    minibatch_size = batch_size // num_minibatches
    num_updates = total_timesteps // batch_size

    processes = []
    for worker_id in range(num_workers):
        p = ctx.Process(target=ppo_mod.ppo_worker, args=(
            worker_id, global_agent, env_id, seed, num_steps, num_updates,
            shared_obs, shared_actions, shared_logprobs, shared_rewards,
            shared_dones, shared_values, shared_next_obs, shared_next_done,
            collect_barrier, update_barrier, ep_queue
        ))
        p.start()
        processes.append(p)

    results = []

    for update in range(num_updates):
        if anneal_lr:
            frac = 1.0 - (update / num_updates)
            lrnow = frac * lr
            optimizer.param_groups[0]["lr"] = lrnow

        collect_barrier.wait()

        with torch.no_grad():
            next_value = global_agent.get_value(shared_next_obs).squeeze(1)
            advantages = torch.zeros_like(shared_rewards)
            lastgaelam = torch.zeros(num_workers)
            for t in reversed(range(num_steps)):
                if t == num_steps - 1:
                    nextnonterminal = 1.0 - shared_next_done
                    nextvalues = next_value
                else:
                    nextnonterminal = 1.0 - shared_dones[t + 1]
                    nextvalues = shared_values[t + 1]
                delta = shared_rewards[t] + gamma * nextvalues * nextnonterminal - shared_values[t]
                lastgaelam = delta + gamma * gae_lambda * nextnonterminal * lastgaelam
                advantages[t] = lastgaelam
            returns = advantages + shared_values

        b_obs = shared_obs.view(-1, obs_dim)
        b_actions = shared_actions.view(-1, 1 if is_discrete else act_dim)
        if is_discrete:
            b_actions = b_actions.squeeze(-1)
        b_logprobs = shared_logprobs.view(-1)
        b_advantages = advantages.view(-1)
        b_returns = returns.view(-1)
        b_values = shared_values.view(-1)

        b_inds = np.arange(batch_size)

        for epoch in range(update_epochs):
            np.random.shuffle(b_inds)
            for start in range(0, batch_size, minibatch_size):
                end = start + minibatch_size
                mb_inds = b_inds[start:end]

                _, newlogprob, entropy, newvalue = global_agent.get_action_and_value(b_obs[mb_inds], b_actions[mb_inds])
                logratio = newlogprob - b_logprobs[mb_inds]
                ratio = logratio.exp()

                with torch.no_grad():
                    approx_kl = ((ratio - 1) - logratio).mean()

                mb_advantages = b_advantages[mb_inds]
                if norm_adv:
                    mb_advantages = (mb_advantages - mb_advantages.mean()) / (mb_advantages.std() + 1e-8)

                pg_loss1 = -mb_advantages * ratio
                pg_loss2 = -mb_advantages * torch.clamp(ratio, 1 - clip_coef, 1 + clip_coef)
                pg_loss = torch.max(pg_loss1, pg_loss2).mean()

                newvalue = newvalue.view(-1)
                if clip_vloss:
                    v_loss_unclipped = (newvalue - b_returns[mb_inds]) ** 2
                    v_clipped = b_values[mb_inds] + torch.clamp(
                        newvalue - b_values[mb_inds],
                        -clip_coef,
                        clip_coef,
                    )
                    v_loss_clipped = (v_clipped - b_returns[mb_inds]) ** 2
                    v_loss_max = torch.max(v_loss_unclipped, v_loss_clipped)
                    v_loss = 0.5 * v_loss_max.mean()
                else:
                    v_loss = 0.5 * ((newvalue - b_returns[mb_inds]) ** 2).mean()

                entropy_loss = entropy.mean()
                loss = pg_loss - ent_coef * entropy_loss + v_loss * vf_coef

                optimizer.zero_grad()
                loss.backward()
                nn.utils.clip_grad_norm_(global_agent.parameters(), max_grad_norm)
                optimizer.step()

            if target_kl is not None and approx_kl > target_kl:
                break

        while not ep_queue.empty():
            results.append(ep_queue.get())

        update_barrier.wait()

    for p in processes:
        p.join()

    return results, global_agent
````

````python
def evaluate_ppo_agent(agent, env_id, episodes=3):
    eval_env = gym.make(env_id)
    returns = []
    for ep in range(episodes):
        obs, _ = eval_env.reset(seed=10_000 + ep)
        done = False
        ep_ret = 0.0
        while not done:
            obs_t = torch.FloatTensor(obs).unsqueeze(0)
            with torch.no_grad():
                if agent.is_discrete:
                    logits = agent.actor_mean(obs_t)
                    action = torch.argmax(logits, dim=1).item()
                else:
                    action = agent.actor_mean(obs_t).squeeze(0).numpy()
            obs, reward, terminated, truncated, _ = eval_env.step(action)
            done = terminated or truncated
            ep_ret += reward
        returns.append(ep_ret)
    eval_env.close()
    return returns


SMOKE_RUNS = 5
SMOKE_TIMESTEPS = max(20_000, ppo_num_workers * ppo_num_steps * 2)
EVAL_EPISODES_PER_RUN = 3

print(f"testing PPO run: runs={SMOKE_RUNS}, timesteps/run={SMOKE_TIMESTEPS}, start_method=spawn")

ppo_smoke_rows = []
for run_idx in range(SMOKE_RUNS):
    run_seed = ppo_seed + 100 * (run_idx + 1)
    t0 = time.time()

    run_results, run_agent = train_ppo(
        env_id=ppo_gym_id,
        seed=run_seed,
        total_timesteps=SMOKE_TIMESTEPS,
        num_workers=ppo_num_workers,
        num_steps=ppo_num_steps,
    )

    train_seconds = time.time() - t0
    eval_returns = evaluate_ppo_agent(run_agent, ppo_gym_id, episodes=EVAL_EPISODES_PER_RUN)

    row = {
        "run": run_idx + 1,
        "seed": run_seed,
        "episode_events": len(run_results),
        "train_seconds": train_seconds,
        "eval_mean": float(np.mean(eval_returns)),
        "eval_std": float(np.std(eval_returns)),
    }
    ppo_smoke_rows.append(row)

    print(
        f"run {row['run']}: seed={row['seed']} | "
        f"events={row['episode_events']} | "
        f"time={row['train_seconds']:.2f}s | "
        f"eval_mean={row['eval_mean']:.2f}"
    )

all_times = np.array([r["train_seconds"] for r in ppo_smoke_rows], dtype=np.float32)
all_eval = np.array([r["eval_mean"] for r in ppo_smoke_rows], dtype=np.float32)
all_events = np.array([r["episode_events"] for r in ppo_smoke_rows], dtype=np.int32)

print("\nppo run summary:")
print(f"time mean/std: {all_times.mean():.2f}s / {all_times.std():.2f}s")
print(f"eval mean/std over runs: {all_eval.mean():.2f} / {all_eval.std():.2f}")
print(f"episode events mean: {all_events.mean():.2f}")
````

````python
def plot_envelope(results_dict, metric_key, title, ylabel):
    plt.figure(figsize=(10, 6))
    colors = ['blue', 'green', 'red', 'purple', 'orange']
    
    for idx, (agent_name, metrics) in enumerate(results_dict.items()):
        if metric_key not in metrics or len(metrics[metric_key]) == 0:
            continue
            
        data = np.array(metrics[metric_key])
        
        min_len = min(len(run) for run in data)
        data = np.array([run[:min_len] for run in data])
        
        if data.size == 0:
            continue
            
        mean_val = np.mean(data, axis=0)
        max_val = np.max(data, axis=0)
        min_val = np.min(data, axis=0)
        x_axis = np.arange(len(mean_val))
        
        plt.plot(x_axis, mean_val, label=agent_name, color=colors[idx % len(colors)])
        plt.fill_between(x_axis, min_val, max_val, color=colors[idx % len(colors)], alpha=0.2)
        
    plt.title(title)
    plt.xlabel('Episodes')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import time
import torch
import torch.optim as optim
import gymnasium as gym

seeds = [10, 20, 30, 40, 50]

def run_pendulum_experiments():
    env_name = 'Pendulum-v1'

    results = {
        'REINFORCE': {'train_rewards': [], 'eval_rewards': [], 'train_time': [], 'wall_clock': [], 'total_steps': []},
        'VPG': {'train_rewards': [], 'eval_rewards': [], 'train_time': [], 'wall_clock': [], 'total_steps': []}
    }

    env_for_steps = gym.make(env_name)
    max_ep_steps = env_for_steps.spec.max_episode_steps if env_for_steps.spec is not None else 200
    env_for_steps.close()

    def train_strategy(net, s):
        out = net(s)
        mean, log_std = torch.chunk(out, 2, dim=-1)
        std = torch.exp(torch.clamp(log_std, min=-20, max=2))
        dist = torch.distributions.Normal(mean, std)
        action = dist.sample()
        log_prob = dist.log_prob(action).sum(dim=-1)
        return action.detach().cpu().numpy()[0], log_prob

    def eval_strategy(net, s):
        out = net(s)
        mean, _ = torch.chunk(out, 2, dim=-1)
        return mean.detach().cpu().numpy()[0], None

    for seed in seeds:
        env = gym.make(env_name)

        agent_reinforce = REINFORCE(
            env=env, seed=seed, gamma=0.99, optimizerFn=optim.Adam, optimizerLR=3e-4,
            MAX_TRAIN_EPISODES=150, MAX_EVAL_EPISODES=10,
            explorationStrategyTrainFn=train_strategy,
            explorationStrategyEvalFn=eval_strategy
        )
        tr, tt, er, wc, _ = agent_reinforce.runREINFORCE()
        results['REINFORCE']['train_rewards'].append(tr)
        results['REINFORCE']['eval_rewards'].append(er)
        results['REINFORCE']['train_time'].append(tt)
        results['REINFORCE']['wall_clock'].append(wc)
        results['REINFORCE']['total_steps'].append((np.arange(len(tr)) + 1) * max_ep_steps)

        agent_vpg = VPG(
            env=env, seed=seed, gamma=0.99, beta=0.01, optimizerFn=optim.Adam, optimizerLR=3e-4,
            MAX_TRAIN_EPISODES=150, MAX_EVAL_EPISODES=10,
            explorationStrategyTrainFn=train_strategy,
            explorationStrategyEvalFn=eval_strategy
        )
        tr, tt, er, wc, _ = agent_vpg.runVPG()
        results['VPG']['train_rewards'].append(tr)
        results['VPG']['eval_rewards'].append(er)
        results['VPG']['train_time'].append(tt)
        results['VPG']['wall_clock'].append(wc)
        results['VPG']['total_steps'].append((np.arange(len(tr)) + 1) * max_ep_steps)

        env.close()

    return results

pendulum_results = run_pendulum_experiments()
plot_envelope(pendulum_results, 'train_rewards', 'Mean Train Rewards (Pendulum-v1)', 'Reward')
plot_envelope(pendulum_results, 'eval_rewards', 'Mean Eval Rewards (Pendulum-v1)', 'Reward')
plot_envelope(pendulum_results, 'train_time', 'Train Time (Pendulum-v1)', 'Time (s)')
plot_envelope(pendulum_results, 'wall_clock', 'Wall Clock Time (Pendulum-v1)', 'Time (s)')
plot_envelope(pendulum_results, 'total_steps', 'Total Steps vs Episodes (Pendulum-v1)', 'Total Steps')
````

````python
def run_halfcheetah_experiments():
    env_name = 'HalfCheetah-v4'

    results = {
        'DDPG': {'train_rewards': [], 'eval_rewards': [], 'train_time': [], 'wall_clock': [], 'total_steps': []},
        'TD3': {'train_rewards': [], 'eval_rewards': [], 'train_time': [], 'wall_clock': [], 'total_steps': []},
        'PPO': {'train_rewards': [], 'eval_rewards': [], 'train_time': [], 'wall_clock': [], 'total_steps': []}
    }

    run_seeds = seeds if 'seeds' in globals() else [0, 1, 2]

    env_for_steps = gym.make(env_name)
    max_ep_steps = env_for_steps.spec.max_episode_steps if env_for_steps.spec is not None else 1000
    env_for_steps.close()

    def eval_ppo_agent(agent, eval_env_name, episodes=5):
        eval_env = gym.make(eval_env_name)
        returns = []
        for ep in range(episodes):
            obs, _ = eval_env.reset(seed=20_000 + ep)
            done = False
            ep_ret = 0.0
            while not done:
                obs_t = torch.FloatTensor(obs).unsqueeze(0)
                with torch.no_grad():
                    if agent.is_discrete:
                        logits = agent.actor_mean(obs_t)
                        action = int(torch.argmax(logits, dim=1).item())
                    else:
                        action = agent.actor_mean(obs_t).squeeze(0).numpy()
                obs, reward, terminated, truncated, _ = eval_env.step(action)
                done = terminated or truncated
                ep_ret += reward
            returns.append(ep_ret)
        eval_env.close()
        return float(np.mean(returns))

    for seed in run_seeds:
        env = gym.make(env_name)

        ddpg_agent = DDPG(
            env, seed=seed, gamma=0.99, tau=0.005, bufferSize=100000, batch_size=128,
            updateFrequency=1, policyOptimizerFn=optim.Adam, valueOptimizerFn=optim.Adam,
            policyOptimizerLR=1e-3, valueOptimizerLR=1e-3, MAX_TRAIN_EPISODES=150,
            MAX_EVAL_EPISODE=10, optimizerFn=optim.Adam
        )
        tr, tt, er, wc, _ = ddpg_agent.runDDPG()
        results['DDPG']['train_rewards'].append(tr)
        results['DDPG']['eval_rewards'].append(er)
        results['DDPG']['train_time'].append(tt)
        results['DDPG']['wall_clock'].append(wc)
        results['DDPG']['total_steps'].append((np.arange(len(tr)) + 1) * max_ep_steps)

        td3_agent = TD3(
            env, gamma=0.99, tau=0.005, bufferSize=100000, updateFrequencyPolicy=2,
            updateFrequencyValue=1, trainPolicyFrequency=2, policyOptimizerFn=optim.Adam,
            valueOptimizerFn=optim.Adam, policyOptimizerLR=1e-3, valueOptimizerLR=1e-3,
            MAX_TRAIN_EPISODES=150, MAX_EVAL_EPISODE=10, optimizerFn=optim.Adam
        )
        tr, tt, er, wc, _ = td3_agent.runTD3()
        results['TD3']['train_rewards'].append(tr)
        results['TD3']['eval_rewards'].append(er)
        results['TD3']['train_time'].append(tt)
        results['TD3']['wall_clock'].append(wc)
        results['TD3']['total_steps'].append((np.arange(len(tr)) + 1) * max_ep_steps)

        ppo_total_timesteps = 150000
        ppo_start = time.time()
        ppo_res, ppo_agent = train_ppo(env_id=env_name, seed=seed, total_timesteps=ppo_total_timesteps)
        ppo_elapsed = time.time() - ppo_start

        if len(ppo_res) > 0:
            _, rewards = zip(*ppo_res)
            rewards = list(rewards)

            ppo_eval_mean = eval_ppo_agent(ppo_agent, env_name, episodes=10)
            eval_series = [ppo_eval_mean for _ in range(len(rewards))]

            ppo_time_series = np.linspace(0.0, ppo_elapsed, num=len(rewards)).tolist()
            ppo_steps_series = np.linspace(0.0, ppo_total_timesteps, num=len(rewards)).tolist()

            results['PPO']['train_rewards'].append(rewards)
            results['PPO']['eval_rewards'].append(eval_series)
            results['PPO']['train_time'].append(ppo_time_series)
            results['PPO']['wall_clock'].append(ppo_time_series)
            results['PPO']['total_steps'].append(ppo_steps_series)
        else:
            ppo_eval_mean = eval_ppo_agent(ppo_agent, env_name, episodes=5)
            results['PPO']['train_rewards'].append([ppo_eval_mean])
            results['PPO']['eval_rewards'].append([ppo_eval_mean])
            results['PPO']['train_time'].append([ppo_elapsed])
            results['PPO']['wall_clock'].append([ppo_elapsed])
            results['PPO']['total_steps'].append([ppo_total_timesteps])

        env.close()

    return results



cheetah_results = run_halfcheetah_experiments()
plot_envelope(cheetah_results, 'train_rewards', 'Mean Train Rewards (HalfCheetah-v4)', 'Reward')
plot_envelope(cheetah_results, 'eval_rewards', 'Mean Eval Rewards (HalfCheetah-v4)', 'Reward')
plot_envelope(cheetah_results, 'train_time', 'Train Time (HalfCheetah-v4)', 'Time (s)')
plot_envelope(cheetah_results, 'wall_clock', 'Wall Clock Time (HalfCheetah-v4)', 'Time (s)')
plot_envelope(cheetah_results, 'total_steps', 'Total Steps vs Episodes (HalfCheetah-v4)', 'Total Steps')
````
